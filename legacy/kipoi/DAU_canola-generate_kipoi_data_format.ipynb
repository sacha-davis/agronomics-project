The purpose of this notebook is to write our canola data into the format that the Kipoi libraries require. This includes a FASTA and an intervals file. 

Please note that this was writted before we knew that there were two organelles in the first dataset, so this generates some outputs that are slightly incorrect.

---

```python
import pandas as pd
```

```python
fasta_out_path = "data/processed/fasta_file"

interval_out_path = "data/processed/intervals_file"

full_data = True
```

```python
df = pd.read_csv('data/raw/CanolaOrganelles_v.1.tsv', sep="\t", header=None)
```

```python
with open(fasta_out_path, "w") as f:  # create pseudo fasta file
    f.write(">chr1\n")
    f.write("".join(df[0].str[0:5])+df.iloc[-1, 0][5:])
```

```python
if full_data:
    # full data
    with open(interval_out_path, "w") as f:  # create invervals file
        for i in df.index:
            f.write("chr1\t"+str(i*5)+"\t"+str(i*5+145)+"\n") # starts at zero
else:
    # shorty
    with open(interval_out_path, "w") as f:  # create invervals file
        for i in range(10):
            f.write("chr1\t"+str(i*5)+"\t"+str(i*5+145)+"\n") # starts at zero
```
